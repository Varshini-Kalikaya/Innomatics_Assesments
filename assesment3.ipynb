```python
#problem 1
employees = {
    "Ravi" : 92,
    "Anita": 88,
    "Kiran" : 92,
    "Suresh" : 85
}
#finding the highest score
highest_score = max(employees.values())
#finding employees with the highest score
top_performers=[name for name, score in employees.items() if score == highest_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (score: {highest_score})")
print("\n" + "-"*50 + "\n")


#problem 2
query = "Buy mobile phone buy phone online"
punctuation = ".,!?;:'\"()[]{}"
for p in punctuation:
    query = query.replace(p, "")
words = query.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
repeated_keywords = {word: count for word, count in word_count.items() if count > 1}
print(repeated_keywords)
print("\n" + "-"*50 + "\n")

#problem 3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
print("\n" + "-"*50 + "\n")

#problem 4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
print("\n" + "-"*50 + "\n")


#problem 5
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
for day, value in enumerate(sales, start=1):
    if value > (average_sales * 1.30):
        print(f"Day {day}: {value}")
print("\n" + "-"*50 + "\n")



#problem 6
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")


```
