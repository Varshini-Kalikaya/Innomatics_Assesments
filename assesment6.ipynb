```python
def smart_parking(capacity,vehicle_logs):
    count=0
    usage=0
    for i in vehicle_logs:
        if i=="IN":
            count+=1
        elif i=="OUT":
            if count>0:
                count-=1
        if count>usage:
            usage=count
        if count>capacity:
            print("Alert:Parkiing capacity exceeded!")
    print("Currently parked vehicles:",count)
    print("Peal Parking usage:",usage)
    if count<capacity:
        print("Parking Status:Available")
    else:
        print("Parking Status:Full")
capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
smart_parking(capacity,vehicle_logs)
```

```python
def estimate_delivery(distance,level,weather):
    #base speed
    speed=20
    #calculate base time in minutes
    base_time=(distance/speed)*60
    #Traffic delays
    if level.lower()=="low":
        traffic_delay=5
    elif level.lower()=="medium":
        traffic_delay=15
    elif level.lower()=="High":
        traffic_delay=25
    else:
        traffic_delay=0
    #weather delays
    if weather.lower()=="clear":
        weather_delay=0
    elif weather.lower()=="rainy":
        weather_delay=6
    elif weather.lower()=="storm":
        weather_delay=20
    else:
        weather_delay=0
    #final ETA
    final_eta=base_time+traffic_delay+weather_delay
    print("Estimated Delivery Time:",round(final_eta),"minutes")
#Function call
estimate_delivery(8,"High","Rainy")
        
    
```

```python
def analyze_seat_booking(total_seats,booked_seats):
    #count booked seats
    booked_count=len(booked_seats)
    #calculate occupancy percentage
    occupancy=(booked_count/total_seats)*100
    print("Occupancy:",round(occupancy),"%")
    #Determine show status
    if occupancy==100:
        print("Show Ststus: Housefull")
    elif occupancy>=75:
        print("Show status:Almost full")
    else:
        print("Show status:Seats available")
    #suggest additional show
    if occupancy>=90:
        print("Suggestion:Consider opening an additonal show.")
total_seats=200
booked_seats=[1]*150
analyze_seat_booking(total_seats,booked_seats)
```

```python
def classify_server_load(cpu_readings):
    sum=0
    #calculate sum
    for i in cpu_readings:
        sum+=i
    avg=sum/len(cpu_readings)
    print("Average CPU Load:",round(avg), "%")
    #classification
    if avg<50:
        print("Server status:Normal")
    elif avg<=80:
        print("Server status:Warning")
    else:
        print("Server status:Critical")
# function call
classify_server_load([45,60,7085,90])
```

```python
def resource_monitor(usage):
    overused=[]
    #loop through dictionary
    for i in usage:
        if usage[i]>8:
            overused.append(i)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert:Yes")
    else:
        print("Overused Resources:None")
        print("Energy Alert:NO")
#function call
usage={"Projector":6,"AC":9,"Lights":4}
resource_monitor(usage)
```

```python
def event_registration(capacity,registration):
    confirmed=0
    waitlisted=0
    #registartion validation
    for i in range(registration):
        if confirmed<capacity:
            confirmed+=1
        else:
            waitlisted+=1
    print("Confirmed Registration:",confirmed)
    print("Waitlisted Users:",waitlisted)
    if confirmed==capacity:
        print("Registration Status:Closed")
    else:
        print("Registration Status:Open")
#function call
event_registration(100,105)
```
