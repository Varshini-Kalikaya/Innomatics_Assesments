```python
def engagement_analyzer(likes):
    total_likes=0
    for like in likes:
        total_likes+=like
    if total_likes>1000:
        status="viral post"
    else:
        status="Normal engagement"
    print(f"Total Likes:{total_likes}")
    print(f"Post Status:{status}")
likes=[400,300,350]
engagement_analyzer(likes)
```

```python
def medical(stock):
    if stock<10:
        status="Low Stock Alert"
    else:
        status="Stock Sufficient"
    print(f"Medicine Stock:{stock}")
    print(f"Status:{status}")
stock=int(input("Enter the medicine stock:"))
medical(stock)
```

```python
def analyze_rainfall(data,required_level):
    total=0
    for rain in data:
        total+=rain
    avg=total/len(data)
    if avg>required_level:
        status="Adequate Rainfall"
    else:
        status="Inaddequate Rainfall"
    print(f"Average Rainfall:{avg}")
    print(f"Rainfall status:{status}")
data=[70,75,80,65]
required_level=70
analyze_rainfall(data,required_level)
```

```python
def duplicate(usernames):
    unique_users=set()
    duplicate_found=False
    for user in usernames:
        if user in unique_users:
            duplicate_found=True
            break
        unique_users.add(user)
    if duplicate_found:
        print("Duplicate Account Found:Yes")
    else:
        print("Duplicate Accounts Found:No")
usernames=["user1","user2","user1","user3"]
duplicate(usernames)
```

```python
def appointment(age):
    if age>=18:
        status="Eligible"
    else:
        status="Not Eligible"
    print(f"Patient age:{age}")
    print(f"Eligibility status:{status}")
age=int(input("Enter the age:"))
appointment(age)
    
```

```python
def agriculture(crop_prices):
    premium_crops=[]
    for price in crop_prices:
        if price>2000:
            premium_crops.append(price)
    print(f"Premium Crops:{premium_crops}")
crop_prices=[1500,2500,1800,3200,1900]
agriculture(crop_prices)
```

```python
def health(error):
    if error==0:
        status="Healthy"
    elif error<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print(f"Error Count:{error}")
    print(f"System status:{status}")
error_count=int(input("Enter the error count:"))
health(error_count)
```

```python
def transaction(amt):
    daily_limit=50000
    if amt<daily_limit:
        status="Approved"
    else:
        status="Rejected"
    print(f"Transaction Amount:{amt}")
    print(f"Transaction status:{status}")
amt=int(input("Enter the amount:"))
transaction(amt)
    
```

```python
def attendence(att):
    if att>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    print(f"Attendence Percentage:{att}")
    print(f"Exam Eligibility:{status}")
att=float(input("Enter the attendence"))
attendence(att)
```

```python
def electricity(units):
    bill=0
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)
    if bill<500:
        status="Low Usage"
    elif bill<=1500:
        status="Moderate Usage"
    else:
        status="High Usage"
    return bill,status
units=int(input("Enter no.of units:"))
bill,status=electricity(units)
print(f"Total Bill: {bill}")
print(f"Usage level:{status}")
```
